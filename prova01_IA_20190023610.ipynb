Os quatro tipos de aprendizados de máquinas são:
- Aprendizagem supervisionada
- Aprendizagem não supervisionada
- Aprendizagem semi-supervisionada
- Aprendizagem por reforço

## A aprendizagem supervisonada: 

A aprendizagem supervisonada consiste em que o sistema recebe dados já rotulados e a saída já é fornecida. Podemos exemplificar esse método como se fosse ensinando matemática para uma criança, onde fornecemos a informação que 2 + 2 = 4, todas as informações sobre o problema é fornecida para que ele possa predeterminar conslusões. Assim, outras somas ela vai conseguir predeterminar formas para se chegar na conclusão. Ou seja, a máquina poderá determinar de forma correta os novos exemplos não rotulados.

Assim, usando a base de dados acima, para o contexto da aprendizagem supervisionada é usada uma porcentagem de mais ou menos 70% de dados, para treinar a máquina com os dados já rotulados, informando suas classes e qual a saída para um determinado conjunto é male ou female. E os outros 30% serve para verificar como que está o desempenho da máquina.  

## A apredizagem não supervisionada

A apredizagem não supervisionada diferente da supersioada, consiste em que o sistema recebe dados não rotulados, predetermindados ou qualquer correlação entre a entrada e a saída. Logo o algoritmo deve procurar uma solução por ele mesmo e não possui dados de referência armazendado. Podemos exemplificar esse método como se fosse uma pessoal que nunca viu futebol ou teve qualquer informação sobre ele, e ela deverá entender o que está acontecendo apenas pelo que está vendo. Logo, a máquina deve observar se encontra algum tipo de padrão entre os dados fornecidos e os agrupar.

Assim, usando a base de dados acima, para o contexto da aprendizagem não supervisionada, é passado os dados para a máquina sem rótulos, ele não vai saber o que cada dado significa, mas vai saber que para este caso ele precisa separar os dados em 2 categorias. A máquina deve procurar padrões nesses dados e os agrupa-los para em duas categorias.

## A aprendizagem semi-supervisionada

A aprendizagem semi-supervisionada é considerada um método intermediário entre a supervisionada e a não supervisionada. Isso por que ela trabalha com informações rotuladas e não rotuladas, ela aprende com os dados rotulados para depois rotular os dados incompletos buscando padrões. 

Assim, usando a base de dados acima, para o contexto de aprendizagem semi-supervisionada, a base terá rotulos faltando, logo será necessário rodar primeiramento em um contexto não supervisionado para que haja o agrupamento de padrões de dados. Isso permitir que os dados rotulados e não rotulados se agrupem de acordo com suas similiaridades nas duas classes de Female e Male, e assim assumimos que esses dados não rotulados sejam classificados de acordo com a classe que estão dentro.

## A aprendizagem por reforço

A aprendizagem por reforço consiste em que o sistema aprenda por meio da tentativa e erro. Nesse método o agente não precisa ser informados de dados previamente para ser treinado como nos outros casos, ele aprende interagindo com o seu ambiente. Geralmente ele funciona com um sistema de recompensa, que toda vez que a máquina atinge um resultado certo ela ganha uma recompensa e quando atinge um resultado errado ela sofre uma penalidade, logo ela busca maximizar os ganhos e minimizar as perdas, pois essa é a medida de desempenho para verificar se suas ações está certa ou não. 

Assim, usar esse método para a base acima é bastante complicado, pois a aprendizagem por reforço tem maior utilização no âmbito mais prático, em contexto de jogos ou em resolver desafios(labirinto). Logo para exemplificar esse tipo de aprendizagem, vamos trazer o exemplo do jogo do dinossauro do goole de quando a não há internet. Nesse jogo as ações possíveis são de pular e abaixar, a máquina e informada disso e que sua medida de desempenho são os pontos obtidos, logo ela vai buscar ações que aumentem esses pontos. Conforme o jogo vai acontecendo ela percebe que ao bater no cacto ou em outro dinossauro, ela perde pontos. Assim, faz com que ela busque formas de evita-los, sendo pulando ou agachando até que ela aprende por tentativa e erro.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

from sklearn import cluster, neighbors, svm, metrics, preprocessing
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1, stratify=y)

test_scores = []
train_scores = []

for i in range(1,15):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
# A melhor Pontuação que vem pelo testes dos dados que usamos no treinamento com diferentes K
max_train_score = max(train_scores)
train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
```

```python
# A melhor Pontuação que vem pelo teste dos dados, que dividimos no ínicio, para teste com diferentes K
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(range(1,15),train_scores,marker='*',label='Train Score')
p = sns.lineplot(range(1,15),test_scores,marker='o',label='Test Score')
```

```python
#Verificar a taxa de erro entre 1<= k < 15
error = []

for i in range(1, 15):  
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 15), error, color='red', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('Valor de K')  
plt.ylabel('Média de erro')  
```

###Pelas informações observadas pelos gráficos, o melhor K para o nosso modelo seria o 7, ele que tem a menor taxa de erro, e nos testes ele que tem a melhor taxa de acerto com o periodo de 1 <= k < 15. Ou seja, sua acurácia é de aproximadamente de 0.93, o qual é uma acurácia bem alta e mehlor do que os outros K's desse período.

O problema dessa base de dados é que a classe Manga tem maior representatividade que as demais classes, ou seja, um problema de desbalanceamento de classes. Isso faz com que o modelo fique induzido em prever objetos mais para a classe Manga do que para as outras classes, podendo acarretar em problemas de na construção de modelos e na geração de previsões. Uma forma de resolver o problema dessa base é utilizar Undersamplim, consiste em reduzir o número de observações da classe majoritária a fim de diminuir essa diferença entre elas. Agora, pode ser feito de duas maneiras, a primeira é com Random Undersampling que é feita uma retirada aleatória da classe, que pode ocasionar um perda grave de informação. E a outra é utilização de unir duas ou mais observações da classe, que ocasiona numa menor perda de informação, que algoritmos de clusterização clássicos podem ser usados, como o KNN.


```python
acuracia = (70+67+55+1000)/(70+10+5+5+20+67+30+4+5+3+55+2+5+20+10+1000)
precisao = (70/(70+20+5+5) + 67/(67+10+3+20) + 55/(55+5+30+10) + 1000/(1000+5+4+2))/4
sensibilidade = (70/(70+10+5+5) + 67/(67+20+30+4) + 55/(55+5+3+2) + 1000/(1000+5+20+10))/4
especificidade = ((67+55+1000)/(67+55+1000+20+5+5) + (70+55+1000)/(70+55+1000+10+3+20) + (70+67+1000)/(70+67+1000+5+30+10) + (70+67+55)/(70+67+55+5+4+2))/4
f1 = 2 * ((precisao * sensibilidade)/(precisao + sensibilidade))

phi1 = ((70*(67+55+1000))-(20+5+5)*(10+5+5))/(math.sqrt((70+20+5+5)*(70+10+5+5)*(67+55+1000+20+5+5)*(67+55+1000+10+5+5)))
phi2 = ((67*(70+55+1000))-(10+3+20)*(20+30+4))/(math.sqrt((67+10+3+20)*(67+20+30+4)*(70+55+1000+10+3+20)*(70+55+1000+20+30+4)))
phi3 = ((55*(70+67+1000))-(5+30+10)*(5+3+2))/(math.sqrt((55+5+30+10)*(55+5+3+2)*(70+67+1000+5+30+10)*(70+67+1000+5+3+2)))
phi4 = ((1000*(70+67+55))-(5+4+2)*(5+20+10))/(math.sqrt((1000+5+4+2)*(1000+5+20+10)*(70+67+55+5+4+2)*(70+67+55+5+20+10)))
phi = (phi1 + phi2 + phi3 + phi4)/4

tvp = sensibilidade
tfp = (1-especificidade)

print("tvp: %0.3f\n" % (tvp))
print("tfp: %0.3f\n" % (tfp))

x = np.array([0, tvp, 1])
y = np.array([0, tfp, 1])

print("Acuracia: %0.3f\n" % acuracia)
print("Precisão: %0.3f\n" % precisao)
print("Sensibilidade: %0.3f\n" % sensibilidade)
print("Especificidade: %0.3f\n" % especificidade)
print("Medida F1: %0.3f\n" % f1)
print("Coeficiente de Mattews: %0.3f\n" % phi)
print("AUC: %0.3f\n" % (np.trapz(x,y)))


```
